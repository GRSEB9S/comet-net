```python
ctrl_age_mean = ctrl_df['age'].mean()
ctrl_age_std  = ctrl_df['age'].std()
print("Mean: {}, Deviation: {}".format(ctrl_age_mean, ctrl_age_std))
```

```python
ctrl_size_mean = ctrl_df['Size [mm]'].mean()
ctrl_size_std  = ctrl_df['Size [mm]'].std()
print("Mean: {}, Deviation: {}".format(ctrl_size_mean, ctrl_size_std))
```

```python
mets_age_mean = mets_df['age'].mean()
mets_age_std  = mets_df['age'].std()
print("Mean: {}, Deviation: {}".format(mets_age_mean, mets_age_std))
```

```python
mets_size_mean = mets_df['Size [mm]'].mean()
mets_size_std  = mets_df['Size [mm]'].std()
print("Mean: {}, Deviation: {}".format(mets_size_mean, mets_size_std))
```

```python
control = ctrl_df['Size [mm]']
metas   = mets_df['Size [mm]']
df_size = pd.DataFrame()
df_size['ctrl'] = control
df_size['mets'] = metas
# df.plot.box()
```

```python
ax = df_size.plot(kind="box", title="Metastasis and Non-Metastasis Tumor Size [mm]")
```

```python
ax = df_size.plot(kind="box",figsize=(9,9), fontsize=16)
ax.set_ylabel("Size [mm]", fontsize=16)
ax.set_title("Metastasis and Non-Metastasis Tumor Size [mm]", fontsize=16)
```
